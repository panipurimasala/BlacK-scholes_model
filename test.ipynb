```python
#importing everything
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import yfinance as yf
import math
```

```python
def black_scholes_call(S, X, T, r, sigma, q):
    d1 = (math.log(S / X) + (r - q + 0.5 * sigma ** 2) * T) / (sigma * math.sqrt(T))
    d2 = d1 - sigma * math.sqrt(T)
    
    N_d1 = 0.5 * (1 + math.erf(d1 / math.sqrt(2)))
    N_d2 = 0.5 * (1 + math.erf(d2 / math.sqrt(2)))
    
    call_price = math.exp(-q * T) * S * N_d1 - math.exp(-r * T) * X * N_d2
    return call_price
```

```python
def dif(stock):
    msft = yf.Ticker(stock)
    date_to_subtract = dt.datetime.strptime(msft.options[0] , '%Y-%m-%d')
    today_date = dt.datetime.now()
    date_difference =  date_to_subtract- today_date
    return date_difference.days
```
